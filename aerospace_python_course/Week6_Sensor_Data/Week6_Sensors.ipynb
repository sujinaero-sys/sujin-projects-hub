```python

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```


## üìä Example 1: IMU Data Simulation

IMU = Inertial Measurement Unit  
It measures **acceleration** and **angular velocity**.


```python

# Time vector
t = np.linspace(0, 10, 200)  # 10 seconds

# Simulated accelerations (m/s^2)
ax = 0.1 * np.sin(0.5*t)
ay = 0.1 * np.cos(0.5*t)
az = 9.81 + 0.05*np.sin(0.2*t)  # gravity + noise

# Simulated gyroscope (rad/s)
gx = 0.01*np.sin(t)
gy = 0.01*np.cos(t)
gz = 0.02*np.sin(0.3*t)

# Plot accelerometer data
plt.plot(t, ax, label="ax")
plt.plot(t, ay, label="ay")
plt.plot(t, az, label="az")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.title("Simulated IMU Accelerometer Data")
plt.legend()
plt.grid(True)
plt.show()
```


## üìç Example 2: GPS Data Simulation

GPS gives latitude, longitude, and velocity.  
We simulate an aircraft path.


```python

# Simulated GPS path
lat0, lon0 = 12.97, 77.59  # Bangalore approx
lat = lat0 + 0.01*np.sin(0.1*t)
lon = lon0 + 0.01*np.cos(0.1*t)
vel = 250 + 5*np.sin(0.2*t)  # velocity (m/s)

plt.plot(lon, lat, label="Flight Path")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Simulated GPS Flight Path")
plt.legend()
plt.grid(True)
plt.show()
```


## ‚õ∞ Example 3: Altimeter Data Simulation

Altimeter measures **altitude vs time**.  
We simulate climb and descent.


```python

altitude = 1000 + 50*t - 2*t**2  # simple parabola climb-descent

plt.plot(t, altitude)
plt.xlabel("Time (s)")
plt.ylabel("Altitude (m)")
plt.title("Simulated Altimeter Data")
plt.grid(True)
plt.show()
```
