```python

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Enable inline plotting (for Jupyter Notebook)
%matplotlib inline
```


## Flight Test Dataset (Simulated)

We‚Äôll simulate a small dataset representing flight parameters: time, altitude, velocity, and thrust.


```python

# Create a simple flight dataset
time = np.arange(0, 11, 1)  # seconds
altitude = 1000 + 50 * time  # meters
velocity = 200 + 5 * time  # m/s
thrust = 100000 - 2000 * time  # Newtons

# Combine into a DataFrame
data = pd.DataFrame({
    'Time (s)': time,
    'Altitude (m)': altitude,
    'Velocity (m/s)': velocity,
    'Thrust (N)': thrust
})

data
```


## Basic Statistical Analysis
We can calculate mean, min, max values to summarize flight performance.


```python

# Summary statistics
data.describe()
```


## Visualization: Altitude vs Time


```python

plt.figure(figsize=(8,5))
plt.plot(data['Time (s)'], data['Altitude (m)'], marker='o')
plt.title("Altitude vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Altitude (m)")
plt.grid(True)
plt.show()
```


## Visualization: Velocity vs Time


```python

plt.figure(figsize=(8,5))
plt.plot(data['Time (s)'], data['Velocity (m/s)'], marker='s', color='orange')
plt.title("Velocity vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.grid(True)
plt.show()
```


## Visualization: Thrust vs Time


```python

plt.figure(figsize=(8,5))
plt.plot(data['Time (s)'], data['Thrust (N)'], marker='^', color='red')
plt.title("Thrust vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Thrust (N)")
plt.grid(True)
plt.show()
```
