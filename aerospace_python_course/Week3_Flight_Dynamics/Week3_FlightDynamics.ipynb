Cell 3 ‚Äî Markdown
## Aerodynamic Forces
- **Lift (L)** = 0.5 √ó œÅ √ó V¬≤ √ó S √ó Cl  
- **Drag (D)** = 0.5 √ó œÅ √ó V¬≤ √ó S √ó Cd  
- **Thrust (T)** = Engine generated force  
- **Weight (W)** = m √ó g  

The balance of these forces determines aircraft motion.


Cell 4 ‚Äî Code

```python
# Parameters
rho = 1.225      # air density (kg/m¬≥)
V = 230          # velocity (m/s)
S = 30.0         # wing area (m¬≤)
Cl = 1.0         # lift coefficient
Cd = 0.03        # drag coefficient
mass = 1200      # kg
g = 9.81         # m/s¬≤

# Calculations
Lift = 0.5 * rho * V**2 * S * Cl
Drag = 0.5 * rho * V**2 * S * Cd
Weight = mass * g
Thrust = Drag  # steady level flight assumption

Lift, Drag, Weight, Thrust
```


Cell 5 ‚Äî Markdown
## Force Visualization
Let's draw a simple diagram showing Lift, Drag, Thrust, and Weight.

Cell 6 ‚Äî Code

```python
plt.figure(figsize=(6,6))
plt.quiver(0,0,0,Lift,angles='xy',scale_units='xy',scale=1,color='blue',label="Lift")
plt.quiver(0,0,Thrust,0,angles='xy',scale_units='xy',scale=1,color='green',label="Thrust")
plt.quiver(0,0,0,-Weight,angles='xy',scale_units='xy',scale=1,color='red',label="Weight")
plt.quiver(0,0,-Drag,0,angles='xy',scale_units='xy',scale=1,color='orange',label="Drag")

plt.xlim(-20000,20000)
plt.ylim(-20000,20000)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Aircraft Force Balance")
plt.show()
```


Cell 7 ‚Äî Markdown
## Longitudinal Stability (Pitching Moment)

Aircraft stability depends on the balance of aerodynamic moments.  
The pitching moment coefficient (Cm) is often expressed as:

**Cm = Cm‚ÇÄ + (dCm/dŒ±) √ó Œ±**  

- Stable aircraft ‚Üí (dCm/dŒ±) < 0  
- Unstable aircraft ‚Üí (dCm/dŒ±) > 0  


Cell 8 ‚Äî Code

```python
# Angle of attack range
alpha = np.linspace(-5, 15, 50)

# Example stability derivative
Cm0 = 0.05
dCm_dAlpha = -0.02  # stable aircraft
Cm = Cm0 + dCm_dAlpha * alpha

plt.figure(figsize=(8,5))
plt.plot(alpha, Cm, label="Pitching Moment Curve")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.xlabel("Angle of Attack (degrees)")
plt.ylabel("Cm")
plt.title("Aircraft Pitching Moment vs Angle of Attack")
plt.legend()
plt.grid(True)
plt.show()
```
